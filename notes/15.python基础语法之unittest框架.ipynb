#### unittest  
一、单元测试框架，unittest（python自带的），pytest(第三方库)  
    编写测试用例---收集测试用例---执行用例---生成测试结果 
    
二、编写测试用例  
1. 创建测试类：类名称建议以Test开头，必须继承unittest.TestCase
2. 测试用例：
    1. 必须写在测试类里面
    2. 函数名称必须是test开头
    3. 执行顺序：按照ASCII码顺序执行
    4. 执行器配置：File---setting---tools---python Integrated Tools ---Testing ---Defalut test runner ---选择unittest
3. 执行
    1. 它不需要初始化
    2. 需要配置执行器

三、测试用例包含哪些东西
1. 前置条件：执行这个场景的前提条件
    1. 函数级别（setUp(self) -> None:）：每个用例执行之前执行一次
    2. 类级别（setUpClass(cls) -> None:）：每个测试类执行之前执行一次，需要重写符
2. 业务逻辑
3. 断言：期望结果与实际结果进行比对
    1. =:对比值
    2. is：对比内存地址
    3. self。assertEqual（期望，实际）：对比是否相等
    如果表达式不成立，就会抛出异常，unittest框架就会根据这个异常去将对应的测试用例标记为失败
4. 后置清理：清理测试数据
    1. 函数级别（tearDown(self) -> None:）：每个用例执行之后执行一次
    2. 类级别（tearDownClass(cls) -> None:）：每个测试类执行之后执行一次，需要重写符
5. 测试报告：收集测试结果

```python
import unittest

#编写测试用例
class TestLogin(unittest.TestCase):
    #前置条件
    def setUp(self) -> None:
        print('每个函数执行之前执行我')

    def tearDown(self) -> None:
        print('每个函数执行完毕后执行我')

    @classmethod
    def setUpClass(cls) -> None:
        print('测试类执行之前执行我')

    def test_login_01(self):
        #业务逻辑
        print('login success1')

    def test_login_02(self):
        print('login success2')

    @classmethod
    def tearDownClass(cls) -> None:
        print('测试类执行之后执行我')
        

```

测试执行  
3. 创建用例运行器  
4. 将测试套件放到运行器执行


```python
import unittest

#编写测试用例
class TestLogin(unittest.TestCase):
    #前置条件
    def setUp(self) -> None:
        print('每个函数执行之前执行我')

    def tearDown(self) -> None:
        print('每个函数执行完毕后执行我')

    @classmethod
    def setUpClass(cls) -> None:
        print('测试类执行之前执行我')

    def test_login_01(self):
        #业务逻辑
        print('login success1')

    def test_login_02(self):
        print('login success2')

    @classmethod
    def tearDownClass(cls) -> None:
        print('测试类执行之后执行我')
        
#实例化/创建测试套件
suite = unittest.TestSuite()

#添加一个测试用例（类（“测试用例名称”））到测试套件中【这个方法不常用】
suite.addTest(TestLogin("test_login_01"))

#创建运行器
#执行通过后用.表示通过
runner = unittest.TextTestRunner()
runner.run(suite)

```

```python
# -*- coding: utf-8 -*-
# @Time: 2021/7/6 15:36
# @Author: xueli
# @File: handleExcel.py
# 对excel中数据进行处理
from openpyxl import load_workbook


class HandleExcel:
    #获取excel中的文件地址、文件名、和sheet名
    def __init__(self,file_path,sheet_name):
        self.file_name = file_path
        self.work_book = load_workbook(filename=self.file_name)
        self.sheet = self.work_book[sheet_name]

    #获取文件中所有测试用例数据
    def get_all_datas(self):
        all_datas = list(self.sheet.iter_rows(values_only = True))
        #获取excel表头和对应数据
        title = all_datas[0]
        case_data = all_datas[1:]
        return title,case_data

    #对所有测试用例数据进行拼接
    def get_case_data_dict(self):
        case_list = []
        title,case_data=self.get_all_datas()
        #将表头和用例数据一一拼接
        for val in case_data:
            res = dict(zip(title,val))
            case_list.append(res)
        return case_list

    #将用例结果写入excel中
    def write_res(self,rows,columns,result=None):
        self.sheet.cell(row=rows,column=columns).value=result
        self.__save_excel()
        self.__close_excel()

    #保存excel
    def __save_excel(self):
        self.work_book.save(filename=self.file_name)
    #关闭excel
    def __close_excel(self):
        self.work_book.clase()
```

```python
# -*- coding: utf-8 -*-
# @Time: 2021/7/7 16:11
# @Author: xueli
# @File: testCasePost.py
# 处理测试用例，进行断言

import unittest
from ddt import ddt,data
import ast

from handleExcel import HandleExcel
from requestMethod import RequestMethod

#实例化测试用例
case_list = HandleExcel(file_path="testCasePost.xlsx",sheet_name='Sheet1').get_case_data_dict()
#实例化接口调用方法
requestMethod = RequestMethod()

@ddt
class TestDemo(unittest.TestCase):
    #获取到所有测试用例
    @data(*case_list)
    def test_Case(self,case):
        #case表示每一条测试用例
        res = requestMethod.user_request_method_post(url=case['url'],data=ast.literal_eval(case['data']))
        #获取实际测试结果
        actual_res={'code':res['code'],'msg':res['msg']}
        # print(actual_res,type(actual_res))
        # print(case['expected'],type(case['expected']))
        # print(type(ast.literal_eval(case['expected'])))
        #将测试结果和实际结果进行对比
        self.assertEqual(first=ast.literal_eval(case['expected']),second=(actual_res))
```
