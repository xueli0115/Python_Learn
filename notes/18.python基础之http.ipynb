#### http协议
一、http协议概念
1. Hyper Text Transfer Protocol(超文本传输协议)
2. 用于从万维网服务器传输文本到本地浏览器的传输协议
3. HTTP协议是基于TCP的应用层协议，它不关心数据传输的细节，主要是用来规定客户端和服务端的数据传输格式，最初是用来向客户端传输HTML页面的内容，默认端口是80
4. HTTP是基于请求与响应模式的、无状态的、应用层的协议

二、HTTP请求类【1~2掌握，3~8了解】
1. get:获取资源
2. post：提交资源
3. put：修改资源
4. delete：删除资源
5. head：查看响应头
6. option：查看支持的请求方式
7. trace：返回服务器收到的请求数据，用于开发自测
8. connect：服务器作为跳板，让服务代理用户去其他网站访问资源，然后再把拿到的资源返回给用户（重定向）

三、 网络架构  
BS架构（浏览器）/CS架构（应用程序app）  
客户端<-------HTTP请求--------> 服务器（前端服务器<----------> 后端服务器（网关、课程服务、鉴权服务））

1. 请求报文
    1. 请求头关注字段：
        1. Accept: */*【用来告知客户端可以处理的内容类型】
        2. token:null【鉴权、后端返回的token信息】
        3. Content-Type:application/json;charset=utf-8【前端告诉后端请求参数编码的格式】
        4. Set-cookie:【前端的缓存信息，session_id（鉴权）会放在这里】
2. 响应报文
    1. 响应头关注字段：
        1. Content-Type:application/json;charset=utf-8
        2. Set-cookie
3. 请求体
    1. get请求：参数拼接在地址后面
    2. post请求：参数放在请求体body里面
    3. get和post区别
        1. 参数传递：get参数在url地址后面，post参数放在body里面
        2. 安全性：get相对不那么安全，post相对安全
        3. 请求参数长度限制（后端可修改）：协议本身并没有做长度限制，是后端服务器做了限制。内存问题
        4. 幂等性：get请求是幂等的，post请求是不幂等（后端要做幂等处理）【幂等：多次请求结果一致】
4. 响应体
    1. get和post无区别

五、HTTP响应状态码  
（一）、1xx：服务器已经收到了请求，需要客户端继续操作
1. 100：客户端应该继续请求
2. 101：切换到更高级的协议

（二）、2xx：HTTP成功状态码
1. 200：请求成功，一般用于get，和post
2. 201:请求已经成功，并创建了新的资源
3. 202：请求已经接受，但未处理完成
4. 204：服务器处理成功，但是并未返回内容
5. 206：服务器成功处理了部分请求，get请求，常用与资源下载

（三）、3xx：资源重定向
1. 301：请求资源被永久移动到新的url，返回替代该资源的url，浏览器会自动重定向到新的url，后续资源的访问将使用新的替代
2. 302：临时移除，301是永久移除，资源只是临时移动到新的url，客户端继续使用原有的url
3. 303：查看该资源的其他地址
4. 304：请求的资源未修改，当服务器返回304，此时服务器不会返回任何资源，客户端将使用本地缓存资源。
5. 305：请求的资源必须通过代理访问
6. 307：临时重定向，与302类似

（四）、4xx：客户端错误
1. 401：请求需要用户的身份认证
2. 402：保留状态码，并未使用
3. 403：服务器已经收到了客户端的请求，但是拒绝执行这个请求
4. 404：服务器无法找到客户端需要访问的资源
5. 407：请求需要代理进行身份认证，与401类似
6. 408：服务器等待客户端发送的请求时间过长，超时

（五）、5xx：服务端错误
1. 500：服务器内部错误，无法完成请求
2. 501：服务器不支持请求的功能，无法完成请求
3. 502：网关或者是代理服务器执行请求的时候，收到了一个无效的响应
4. 503：服务器故障，暂时无法处理客户端的请求
5. 504：网关或者是服务器代理，无法从远处服务器获取请求
6. 505：服务器不支持请求的HTTP协议的版本，无法处理请求

4. get请求

```python
#get请求方式一
url_01 = 'http://httpbin.org/get?key1=val1&key2=val2'
res_01 = requests.get(url = url_01)
pprint(res_01.json())

#get请求方式二
url_02 = 'http://httpbin.org/get'
data_02 = {'key1':'val1','key2':'val2'}
res_02 = requests.get(url = url_02,params=data_02)
pprint(res_02.json())
```

5. post请求
    1. data
        1. 默认使用'Content-Type':'application/x-www-form-urlencode'
        2. 请求头指定了Content-Type类型，就会使用指定的类型
    2. json
        1. 默认使用'Content-Type':'application/json'
        2. 请求头指定了Content-Type类型，就会使用指定的类型 

```python
#post请求方法一
url_03 = 'http://httpbin.org/post'
data_03 = {'key1':'val1','key2':'val2'}
res_03 = requests.post (url = url_03,json=data_03)
pprint(res_03.json())

#post请求方法二
header_03 ={'Content-Type':'application/form-data'}
url_03 = 'http://httpbin.org/post'
data_03 = {'key1':'val1','key2':'val2'}
res_03 = requests.post (url = url_03,json=data_03,headers =header_03)
pprint(res_03.json())
```
