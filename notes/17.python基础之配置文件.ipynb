2. ini文件格式  
[section1]  
key1=val1  
key2=val2  
[section2]  
key1=val1  
key2=val2  

{section:{key1:val1,key2:val2},section:{key1:val1,key2:val2}}

书写注意：  
1. 等号左右两边不能有空格
2. section不能重复
3. 读取数据默认是字符串类型


```python
#test_ini.ini文件
[mysql]
host=127.0.0.1
port=3306
[log]
dirpath=val1
level=val2
```

```python
# -*- coding: utf-8 -*-
# @Time: 2021/7/9 15:36
# @Author: xueli
# @File: conf_ini.py


from configparser import ConfigParser
import os

conf = ConfigParser()
#读文件
conf.read(filenames='test_ini.ini',encoding='utf-8')
# filename = os.path.dirname(__file__)
# conf.read(filenames=filename,encoding='utf-8')

#获取所有的section
sect = conf.sections()
print(sect)

# sect_02 = conf.keys()
# print('这里获取的是对象：',sect_02)
# for i in sect_02:
#     print(i)

#获取对应sections下所有的key
keys = conf.options("mysql")
print(keys)

#获取对应sections下所有的key和value
val = conf.items('mysql')
print(val)

#获取某个section中某个key的dval
val_val = conf.get(section='mysql',option='host')
print(val_val)

port_val = conf.getint(section='mysql',option='port')
print(port_val)
```

四、封装
1. 方法一

```python
# -*- coding: utf-8 -*-
# @Time: 2021/7/15 16:07
# @Author: xueli
# @File: handle_ini.py

from configparser import ConfigParser

class HandleIni:

    def __init__(self,filename):
        self.conf = ConfigParser()
        self.conf.read(filenames=filename,encoding='utf-8')

    def get_str_val(self,section):
        # 获取对应sections下所有的key和value
        val = self.conf.items(section=section)
        return val

    def get_str_val_val(self,section,optioon):
        # 获取某个section中某个key的dval
        val_val = self.conf.get(section=section, option=optioon)
        return  val_val

    def get_str_port_val(self,section,optioon):
        port_val = self.conf.getint(section=section, option=optioon)
        return port_val

if __name__=="__main__":
    cl = HandleIni(filename='test_ini.ini')
    port = cl.get_str_port_val(section='mysql',optioon='port')
    print(port)
```
